### Challenge 3: Hidden Cold Joke

```python
response = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents')
results = response.json()
results
```

```python
def scavengerhunt():
    data_js = []
    for result in results:
        if result["name"] == '.gitignore':
            pass
        else:
            datas = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}'.format(result["name"]))
            data_js += datas.json()
    return data_js

data_js = scavengerhunt()
data_js
```

```python
def dataaaaa():
    lista = {}
    for bo in data_js:
        if bo["name"].endswith('.scavengerhunt'):
            lista[bo["name"]] = bo["download_url"]
    return lista

lista = dataaaaa()
lista
```

```python
def ordedic():
    dic ={}
    for key in sorted(lista):
        dic[key] = lista[key]
    return dic

dic = ordedic()
dic
```

```python
def cadena():
    lista = []
    for key in dic:
        lista.append(requests.get(dic[key]).text[:-1])
    cadena = " ".join(lista)
    return cadena

cadena()
```
